```python
import pandas as pd
import os
import re
import collections
```

```python
cosmetic = pd.read_csv('../dataset/cosmetics_reviews.csv')
```

```python
cosmetic
```

```python
reviews = cosmetic['review']
```

```python
reviews
```

##### 특수문자 & 이모티콘 제거

```python
result = []
for review in reviews:
    review = re.sub('[n\{\}\[\]\/?.,;:|\)*~`!^\-_+<>@\#$%&\\\=\(\'\"ㅠㅋㅜㅇㅎㅡ\n]', ' ', review).strip()
    emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"    # emoticons
        u"\U0001F300-\U0001F5FF"    # symbols & pictographs
        u"\U0001F680-\U0001F6FF"    # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"    # flags (iOS) 
                               "]+", flags=re.UNICODE)
    review = emoji_pattern.sub(r'', review)
    result.append(review)
```

```python
len(result)
```

```python
result[102]
```

```python
cosmetic['review'] = result
```

##### 앞뒤 공백 제거

```python
result = []
for review in reviews:
    review = review.strip()
    result.append(review)
```

```python
cosmetic['review'] = result
```
