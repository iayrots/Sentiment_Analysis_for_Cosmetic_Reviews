```python
# 수집한 리뷰데이터를 okja라는 객체에 저장
okja = []
file = open('okja.txt', 'r', encoding='utf-8')

lines = file.readlines()
for line in lines:
    okja.append(line)

file.close()
```

```python
okja
```

```python
len(okja)
```

```python
# KoNLPy 모듈을 사용해서 sentences_tag라는 리스트 형태로 저장
from konlpy.tag import Twitter

twitter = Twitter()

sentences_tag = []
for sentence in okja:
    sentences_tag.append(twitter.pos(sentence)) # pos : 품사 부착
```

```python
sentences_tag
```

```python
# token 갯수
len(sentences_tag)
```

```python
# 자주 사용하는 단어 추출 (명사와 형용사 token만 추출)
noun_adj_list = []
for sentence in sentences_tag:
    for word, tag in sentence:
        if tag in ['Noun', 'Adjective']:
            noun_adj_list.append(word)
```

```python
# collections 모듈을 이용해서 most frequent word 10가지 출력
from collections import Counter
counts = Counter(noun_adj_list)
counts.most_common(10)
```
