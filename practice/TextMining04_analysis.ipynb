### 영화 리뷰 간 유사성 계산

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

##### - '옥자' 리뷰를 doc1에 저장

```python
review_list = []
for n in range(30):
    url = 'http://movie.daum.net/moviedb/grade?movieId=97728&type=netizen&page={}'.format(n+1)
    webpage = urlopen(url)
    source = BeautifulSoup(webpage, 'html.parser', from_encoding='utf-8')
    reviews = source.findAll('p', {'class':'desc_review'})
    
    for review in reviews:
        review_list.append(review.get_text().strip().replace('\n','').replace('\t','').replace('\r',''))
        
```

```python
len(review_list)
```

```python
file = open('okja.txt', 'w', encoding='utf-8')
for review in review_list:
    file.write(review + '\n')
file.close()
```

```python
doc1 = ''
file = open('okja.txt', 'r', encoding='utf-8')
lines = file.readlines()
for line in lines:
    doc1 += line
file.close()
```

```python
doc1
```

##### - '설국열차' 리뷰를 doc2에 저장

```python
review_list = []
for n in range(30):
    url = 'http://movie.daum.net/moviedb/grade?movieId=42238&type=netizen&page={}'.format(n+1)
    webpage = urlopen(url)
    source = BeautifulSoup(webpage, 'html.parser', from_encoding='utf-8')
    reviews = source.findAll('p', {'class':'desc_review'})
    
    for review in reviews:
        review_list.append(review.get_text().strip().replace('\n','').replace('\t','').replace('\r',''))
        
        
file = open('sulkuk.txt', 'w', encoding='utf-8')
for review in review_list:
    file.write(review + '\n')
file.close()


doc2 = ''
file = open('sulkuk.txt', 'r', encoding='utf-8')
lines = file.readlines()
for line in lines:
    doc2 += line
file.close()
```

```python
doc2
```

##### - '어벤져스:인피니티 워' 리뷰를 doc3에 저장

```python
review_list = []
for n in range(30):
    url = 'http://movie.daum.net/moviedb/grade?movieId=93251&type=netizen&page={}'.format(n+1)
    webpage = urlopen(url)
    source = BeautifulSoup(webpage, 'html.parser', from_encoding='utf-8')
    reviews = source.findAll('p', {'class':'desc_review'})
    
    for review in reviews:
        review_list.append(review.get_text().strip().replace('\n','').replace('\t','').replace('\r',''))
        
        
file = open('avengers.txt', 'w', encoding='utf-8')
for review in review_list:
    file.write(review + '\n')
file.close()


doc3 = ''
file = open('avengers.txt', 'r', encoding='utf-8')
lines = file.readlines()
for line in lines:
    doc3 += line
file.close()
```

```python
doc3
```

##### TF-IDF로 벡터화 & 코사인 유사도 (세 리뷰의 명사들간의 유사성 확인)

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
corpus = [doc1, doc2, doc3]
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(corpus)
X = X.todense()
```

```python
print("similarity between 'okja' and 'sulkuk':", cosine_similarity(X[0],X[1]))
print("similarity between 'okja' and 'avengers':", cosine_similarity(X[0],X[2]))
print("similarity between 'sulkuk' and 'avengers':", cosine_similarity(X[1],X[2]))
```
