```python
import tweepy
import settings as s

# 트위터 Application에서 발급 받은 key 정보들
consumer_key = s.APP_KEY
consumer_secret = s.APP_SECRET
access_token = s.OAUTH_TOKEN
access_token_secret= s.OAUTH_TOKEN_SECRET

# 1. 핸들러 생성 및 개인정보 인증요청
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)

# 2. 액세스 요청
auth.set_access_token(access_token, access_token_secret)

# 3. twitter API 생성
api = tweepy.API(auth)

keyword = "고맙다 OR 기쁘다 OR 좋다 OR 만족하다 OR 행복하다" # 검색하고 싶은 키워드 입력 OR로 연결
result = [] # 크롤링 텍스터를 저장할 리스트 변수

for i in range(1, 15): # 1,2 페이지 크롤링
    # keyword 검색 실시, 결과가 tweets 변수에 담긴다
    tweets = api.search(keyword)  
    for tweet in tweets:
        # 크롤링 결과 리스트에 삽입(id, 트윗내용, 생성날짜)
        result.append([tweet.id_str, tweet.text, tweet.created_at]) 
        
```

```python
print(len(result)) # 크롤링하여 가져온 트윗 개수
print(result) # 전체 결과 확인
```
